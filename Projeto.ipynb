```python
from random import randint

lista = ['banana', 'melão', 'maça', 'sabão', 'jaca', 'salgadinho', 'uva', 'pera', 'melancia', 'pasta']
precos = []

for preco in lista:
    precos.append(randint(3, 30))

usuario = input('Qual função você deseja fazer? ')
if usuario == 'adicionar':
    for i in lista:
        produto = input('Informe o produto que queira adicionar: ').lower().strip()
        if produto == 'parar':
            break
        lista.append(produto)
        preco = randint(3, 30)
        precos.append(preco)
        print(f'Preços dos produtos: {lista}')
            
elif usuario == 'remover': 
    for i in lista:
        remover = input('Informe o produto que você queira excluir: ').lower().strip()
        if remover == 'parar':
            break
        if remover in lista:
            index = lista.index(remover)
            lista.pop(index)
            precos.pop(index)
        else:
            print("Produto não encontrado.")
        
        print(f'Lista atualizada de Produtos: {lista}')
else:
    print('Função não existente')
```
